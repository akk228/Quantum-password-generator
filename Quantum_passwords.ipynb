```python
################################ FUNCTIONS ###########################################
# This fucntion generates random number based on a single measurement of quantum state
# and then returns reminder from division by Mod

def quantum_random_number(N, Mod):
    from qiskit import QuantumCircuit, Aer, assemble
    # Assemble quantum circuit from homogeneous superposition of N qubits
    # This circuit create tensot product of hadamard states which yields to all the states
    # being distributed homogeneously, and each of which represents a base two number
    qc = QuantumCircuit(N)
    for q in range(N):
        qc.h(q)
    qc.measure_all()
    
    # Run quantum circuit on a simulator of quantum chip
    # And get a measured state which will corespond to some binary number
    
    # Do not modify this part to run on real device,
    # otherwise, u'll kill yourself before u get your turn in the queue for running your algorithm
    
    simulation = Aer.get_backend('aer_simulator')
    results = simulation.run(assemble(qc), shots=1).result().get_counts()
    
    # Extract the result of measurement as a base 2 number and converts it to the decimal number
    
    random_number = int(list(results)[0],2)
    # I use the reminder from division by Mod because I use slightly more qubits than it's necessary
    return random_number % Mod

# Password generator

def password_generator ():
    from numpy import log2, floor
    import string
    
    print('Would you like to generate password automatically?\n\n')
    
    replies=['a','m']
    reply = '' 
    
    while reply not in replies:
        reply = input('Press "a" for automatic generation\nPress "m" for manual generation\n')
        if reply not in replies:
            print('Error, press "a" or "m": ')
    
    if reply == 'a':
        while True:
            # I need to add here exception handling in case inputed length is not an integer number
            length = int(input('Choose length os password: '))
            if length > 4:
                break
            else:
                print("Password can't be less than 5 characters, try more.")
        
        alphabet = string.ascii_letters + string.digits + string.punctuation
        
        #generate password of size = length using generator of quantum numbers
        
        print('Number of qubits for number generation is '+ str(int(floor(log2(len(alphabet))))) )
        
        return ''.join(alphabet[ quantum_random_number(1+int(floor(log2(len(alphabet)))), len(alphabet))] for i in range(length))
    
    elif reply == 'm':
        while True:
            # Add here exception handling in case inputed length is not an integer number
            length = int(input('Choose length os password: '))
            if length > 2:
                break
            else:
                print("Password can't be less than 3 characters, try more.")
        
        replies_2 = ['y','n']
        
        while True:
            letters = input('Letters in password? y/n: ')
            while letters not in replies_2:
                letters = input('Error, enter "y" or "n": ')
            
            digits = input('Digits in password? y/n: ')
            while digits not in replies_2:
                digits = input('Error, enter "y" or "n": ')
        
            spc_char = input('Special characters in password? y/n: ')
            while spc_char not in replies_2:
                spc_char = input('Error, enter "y" or "n": ')
            
            if 'y' in (letters,digits,spc_char):
                break
            else:
                print('At least one kind of symbols has to be present in password content\nTry again.\n')
            
        alphabet = ''
            
        if letters == 'y':
            alphabet += string.ascii_letters
        
        if digits == 'y':
            alphabet += string.digits
            
        if spc_char == 'y':
            alphabet += string.punctuation
        
        return ''.join(alphabet[ quantum_random_number(1+int(floor(log2(len(alphabet)))), len(alphabet))] for i in range(length))
```
